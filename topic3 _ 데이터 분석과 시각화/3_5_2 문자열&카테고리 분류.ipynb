## 실습과제 _ 박물관이 살이있다 1

한국에서 잘나가는 동양예술전문가 솔희는 최근 “박물관이 살아 있다” 프로젝트를 시작했습니다.

“박물관이 살아 있다” 프로젝트는 점점 떨어져가는 문화예술공간의 방문율을 높이기 위해 시작되었습니다.
김솔희씨는 먼저 예술의 흥행을 위해선 젊은이들의 참여가 시급하다고 판단하여, 대학교 박물관을 먼저 개선하기로 하였습니다.

대학 박물관을 개선하기 위해 다음과 같이 박물관을 분류하기로 하였습니다.

- 박물관은 대학/일반 박물관으로 나뉜다.
- 시설명에 '대학'이 포함되어 있으면 '대학', 그렇지 않으면 '일반'으로 나누어 '분류' column에 입력한다.

'분류' column을 만들어서 솔희를 도와주세요!



```python
import pandas as pd

df = pd.read_csv('C:/Users/jaewoo/Desktop/data science/data/museum_1.csv')

# 분류 컬럼 추가
df['분류'] = '일반'

# 시설명에 대학을 포함하고 있는 index 출력
univ = df[df['시설명'].str.contains('대학')].index

# 대학을 포함한 index의 값을 대학으로 변경
df.loc[univ, '분류']= '대학'
df
```

'지역번호' column을 '지역명' 으로 변경하고, 아래 규칙에 따라 지역을 넣어주세요.

```python
%matplotlib inline
import pandas as pd


df = pd.read_csv('C:/Users/jaewoo/Desktop/data science/data/museum_3.csv', dtype={'지역번호': str})
df
```

```python
local_number = {
    '02' : '서울시',
    '031' : '경기도',
    '032' : '경기도',
    '033' : '강원도',
    '041' : '충청도',
    '042' : '충청도',
    '043' : '충청도',
    '044' : '충청도',
    '051' : '부산시',
    '052' : '경상도',
    '053' : '경상도',
    '054' : '경상도',
    '055' : '경상도',
    '061' : '전라도',
    '062' : '전라도',
    '063' : '전라도',
    '064' : '제주도',
    '1577' : '기타',
    '070' : '기타'
}
df['지역명'] = df['지역번호'].map(local_number)
df.drop('지역번호', axis=1, inplace=True)
df
```

## 실습과제 _ 직업탐구하기 2

실습과제
이번에는 여자 비율이 높은 직업과, 남자 비율이 높은 직업이 무엇인지 궁금한데요.

groupby 문법을 사용해서 '여성 비율'이 높은 순으로 직업을 나열해 보세요.

DataFrame이 아닌 Series로, 'gender'에 대한 값만 아래와 같이 출력되어야 합니다.

```python
df['gender'][df['gender']=='F'] = 1
df['gender'][df['gender']=='M'] = 0

df['gender'] = df['gender'].astype(int)
gender = df.groupby('occupation')
gender.mean()['gender'].sort_values(ascending = False)
```
