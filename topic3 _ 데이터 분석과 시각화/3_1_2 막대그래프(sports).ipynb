실리콘 밸리에서 일하는 남자 관리자 (Managers)에 대한 인종 분포를 막대 그래프로 다음과 같이 그려보세요.

```python
%matplotlib inline
import pandas as pd

# 실리콘 밸리에서 일하는 남자 관리자(Manager)에 대한 인종 분포를 막대 그래프로 그려보세요
df = pd.read_csv('C:/Users/jaewoo/Desktop/data science/data/silicon_valley_summary.csv', index_col = 0)
df
```

```python
male_manager = (df['gender'] == 'Male') & (df.index == 'Managers')
df1 = df.loc[male_manager, ['race_ethnicity', 'count']]
df1
```

```python
df1.set_index('race_ethnicity', inplace = True)
df1.drop('All', axis= 'index', inplace = True)
df1['count'].plot(kind='bar')
```

```python
data = df[(df.index == 'Managers') & (df['gender'] =='Male') & (df['race_ethnicity'] != 'All')]
data.plot(kind='bar', x='race_ethnicity', y='count')
```
