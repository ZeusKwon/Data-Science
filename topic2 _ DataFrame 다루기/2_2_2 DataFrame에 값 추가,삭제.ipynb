# DataFrame Row, Column 추가/삭제

```python
import pandas as pd
iphone_df = pd.read_csv('C:/Users/jaewoo/Desktop/data science/data/iphone.csv', index_col = 0)
iphone_df
```

## 값 추가

```python
iphone_df.loc['iPhone XR'] = ['2018-1026', 6.1, '3GB', 'iOS 12.0.1', 'Yes']
iphone_df
```

```python
iphone_df['제조사'] = 'Apple'
iphone_df
```

## 값 삭제

- 기존 데이터는 변경하지 않고 Row 삭제

```python
iphone_df.drop('iPhone XR', axis = 'index')
iphone_df
```

- 기존데이터도 변경하고 Row 삭제

```python
iphone_df.drop('iPhone XR', axis = 'index', inplace = True)
iphone_df
```

- 기존데이터도 변경하고 Column 삭제

```python
iphone_df.drop('제조사', axis = 'columns', inplace = True)
iphone_df
```

- 여러 Row 삭제

```python
iphone_df.drop(['iPhone 7', 'iPhone 8', 'iPhone X'], axis = 'index', inplace = True)
iphone_df
```
